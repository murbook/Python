```python
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.preprocessing import image
import numpy as np
from IPython.core.display import Image, display
```

```python
vgg16 = VGG16(weights='imagenet')
```

```python
def predict(img_file_path):
    display(Image(img_file_path, width=150, unconfined=True))
    img = image.load_img(img_file_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    preds = vgg16.predict(preprocess_input(x))

    # 予測確率が高いトップ5を出力
    top = 5
    results = decode_predictions(preds, top=top)[0]
    for result in results:
        print(result)
```

```python
predict('2016-06-14_Orange_and_white_tabby_cat_born_in_2016_茶トラ白ねこ_DSCF6526☆彡.jpg')
```
